```python
psqi_related_columns = df.columns[57:76].tolist()

df_psqi = df[psqi_related_columns].copy()

#df['2. How long (in minutes) has it taken you to fall asleep each night?'].unique()
```

These match up with standard PSQI components:


PSQI Component	Related Columns You Have

4. Sleep Disturbances	7–15

```python
# Example categories to ordinal mapping
q2_column = psqi_related_columns[1]  # Index 1 in your PSQI columns list

# Clean and standardize
df_psqi[q2_column] = df_psqi[q2_column].str.strip().str.lower()

# Define ordinal mapping for Q2 categories
q2_mapping = {
    '≤ 15 minutes': 0,
    'less than 15 minutes': 0,
    '15 minutes': 0,
    '16-30 minutes': 1,
    '16 – 30 minutes': 1,
    'about 20 minutes': 1,
    '31-60 minutes': 2,
    'more than 30 minutes': 2,
    '> 60 minutes': 3,
    'more than 1 hour': 3,
    'more than 60 minutes': 3
}

# Apply ordinal encoding
df_psqi['Q2_score'] = df_psqi[q2_column].map(q2_mapping)

# Handle any unmapped values (optional)
df_psqi['Q2_score'] = df_psqi['Q2_score'].fillna(0)  # or np.nan if you want to keep missing

# Check results
print(df_psqi[[q2_column, 'Q2_score']].head())
```

### Encoding Q6

```python
# Q6 column
q6_column = psqi_related_columns[5]  # Index 5 in your PSQI columns list

# Clean Q6 values
df_psqi[q6_column] = df_psqi[q6_column].str.strip().str.lower()

# Define ordinal mapping for Q6
q6_mapping = {
    'not during the past month': 0,
    'less than once a week': 1,
    'once or twice a week': 2,
    'three or more times a week': 3
}

# Apply ordinal encoding
df_psqi['Q6_score'] = df_psqi[q6_column].map(q6_mapping)

# Handle unmapped values
df_psqi['Q6_score'] = df_psqi['Q6_score'].fillna(0)  # or np.nan if needed

# Check results
print(df_psqi[[q6_column, 'Q6_score']].head())
```

```python
# Combine Q2 and Q6 scores
df_psqi['C2_sum'] = df_psqi['Q2_score'] + df_psqi['Q6_score']

# Map sum to Component 2 score
def component2_score(total):
    if total == 0:
        return 0
    elif 1 <= total <= 2:
        return 1
    elif 3 <= total <= 4:
        return 2
    else:
        return 3

df_psqi['C2'] = df_psqi['C2_sum'].apply(component2_score)

# Check results
print(df_psqi[['Q2_score', 'Q6_score', 'C2_sum', 'C2']].head())
```

### Sleep Duration

```python
# Q4 column
q4_column = psqi_related_columns[3]  # Index 3 in your PSQI columns list

# Clean values
df_psqi[q4_column] = df_psqi[q4_column].str.strip().str.lower()

# Map Q4 to numeric values for efficiency calculation
q4_numeric_mapping = {
    'more than 8 hours': 8.5,
    '7-8 hours': 7.5,
    '4-6 hours': 5,
    'less than 4 hours': 3
}

df_psqi['Q4_numeric'] = df_psqi[q4_column].map(q4_numeric_mapping)

# Check results
print(df_psqi[[q4_column, 'C3']].head())

```

```python
q5_column = psqi_related_columns[4]  # Index 4 in your PSQI columns list

# Let's check Q5 values
print(df_psqi[q5_column].unique())
```

```python
# Clean and standardize Q5 values
df_psqi[q5_column] = df_psqi[q5_column].str.strip().str.lower()

# Map Q5 (Hours in Bed) to numeric values
q5_numeric_mapping = {
    'more than 10 hours': 10.5,
    '8-10 hours': 9,
    '5-7 hours': 6,
    'less than 5 hours': 4
}

df_psqi['Q5_numeric'] = df_psqi[q5_column].map(q5_numeric_mapping)
# Check results
print(df_psqi[[q5_column, 'Q5_numeric']].head())
```

```python
# Calculate sleep efficiency
df_psqi['efficiency'] = (df_psqi['Q4_numeric'] / df_psqi['Q5_numeric']) * 100
# Check results
print(df_psqi[['Q4_numeric', 'Q5_numeric', 'efficiency']].head())

```

```python
# Scoring Sleep Efficiency
def efficiency_score(eff):
    if eff > 85:
        return 0
    elif 75 <= eff <= 84:
        return 1
    elif 65 <= eff <= 74:
        return 2
    else:
        return 3

df_psqi['C4'] = df_psqi['efficiency'].apply(efficiency_score)

# Check results
print(df_psqi[['Q4_numeric', 'Q5_numeric', 'efficiency', 'C4']].head())
```

### Component 5 Sleep Disturbances.
 Questions Involved (Q7–Q15):
Waking up at night, bathroom use, snoring, temperature discomfort, bad dreams, pain, etc.

🎯 Scoring for Each Q7–Q15:
0 points → Not during the past month

1 point → Less than once a week

2 points → Once or twice a week

3 points → Three or more times a week

🔹 Component 5 Calculation:
Take sum of Q7–Q15 scores, then:

0 points → Sum = 0

1 point → Sum = 1–9

2 points → Sum = 10–18

3 points → Sum = 19+

```python
# Q7–Q15 columns
disturbance_cols = psqi_related_columns[6:15]  # Index 6 to 14

# Clean values for Q7–Q15
df_psqi[disturbance_cols] = df_psqi[disturbance_cols].apply(lambda col: col.str.strip().str.lower())

# Define ordinal mapping for all disturbance questions
disturbance_mapping = {
    'not during last month': 0,
    'less then once a week': 1,
    'once or twice a week': 2,
    'three or more in week': 3
}
# ['not during last month' 'three or more in week' 'less then once a week'
#  'once or twice a week']
# Apply mapping
for col in disturbance_cols:
    df_psqi[col + '_score'] = df_psqi[col].map(disturbance_mapping)

# Sum the disturbance scores
disturbance_score_cols = [col + '_score' for col in disturbance_cols]
df_psqi['disturbance_sum'] = df_psqi[disturbance_score_cols].sum(axis=1)
```

```python
# Component 5 scoring
def disturbance_component_score(total):
    if total == 0:
        return 0
    elif 1 <= total <= 9:
        return 1
    elif 10 <= total <= 18:
        return 2
    else:
        return 3

df_psqi['C5'] = df_psqi['disturbance_sum'].apply(disturbance_component_score)

# Check results
print(df_psqi[['disturbance_sum', 'C5']].head())
```

### Day Time Disfunctioning.
 Component 7 (C7): Daytime Dysfunction
🔹 Questions:
Q17: "In the last month, how many times you cannot sleep due to any program or other important case?"

Q18: "In the last month, how many times you face problems to maintain program or other important case?"

🎯 Scoring for Q17 & Q18:
0 points → Not during the past month

1 point → Less than once a week

2 points → Once or twice a week

3 points → Three or more times a week

🔹 Component 7 Logic:
Sum of Q17_score + Q18_score → Map to C7:

0 points → Sum = 0

1 point → Sum = 1–2

2 points → Sum = 3–4

3 points → Sum = 5–6



```python
# Q17 & Q18 columns
q17_column = psqi_related_columns[16]
q18_column = psqi_related_columns[17]

# Clean values
df_psqi[q17_column] = df_psqi[q17_column].str.strip().str.lower()
df_psqi[q18_column] = df_psqi[q18_column].str.strip().str.lower()

# Let's check their unique values
print(df_psqi[q17_column].unique())
print(df_psqi[q18_column].unique())
```

```python
# Define mapping
daytime_mapping = {
    'not during last month': 0,
    'less then once a week': 1,
    'once or twice a week': 2,
    'three or more in week': 3
}

# Map Q17 & Q18
df_psqi['Q17_score'] = df_psqi[q17_column].map(daytime_mapping)
df_psqi['Q18_score'] = df_psqi[q18_column].map(daytime_mapping)

# Sum for Component 7
df_psqi['daytime_sum'] = df_psqi['Q17_score'] + df_psqi['Q18_score']
```

```python
# Component 7 scoring
def dysfunction_score(total):
    if total == 0:
        return 0
    elif 1 <= total <= 2:
        return 1
    elif 3 <= total <= 4:
        return 2
    else:
        return 3

df_psqi['C7'] = df_psqi['daytime_sum'].apply(dysfunction_score)

# Check results
print(df_psqi[['Q17_score', 'Q18_score', 'daytime_sum', 'C7']].head())
```
